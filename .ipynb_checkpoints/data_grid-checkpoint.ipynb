```python
import csv


def create_grid(x_range, y_range):
    num_cells = 10
    # dictionary of cells
    cells = dict()
    min_x = x_range[0]
    max_x = x_range[1]
    min_y = y_range[0]
    max_y = y_range[1]
    
    # Create a grid of cells with equal range of values
    for x in range(num_cells):
        for y in range(num_cells):
            cells[(x, y)] = [[min_x + x*(max_x - min_x)/10, min_y + y*(max_y - min_y)/10], [min_x + (x+1)*(max_x - min_x)/10, min_y + (y+1)*(max_y - min_y)/10]]
    return cells


def add_mbr_to_grid(records, grid):
    # check if MBR is in cell
    for road in records:
        min_mbr_x = road[1][0][0]
        max_mbr_x = road[1][1][0]
        min_mbr_y = road[1][0][1]
        max_mbr_y = road[1][1][1]
        for key in grid:
            grid_coordinates = grid[key]
            if min_mbr_x >= grid_coordinates[0][0] and max_mbr_x <= grid_coordinates[1][0] and min_mbr_y >= grid_coordinates[0][1] and max_mbr_y <= grid_coordinates[1][1]:
                grid[key].append(road[0])
                print("added road with id:", road[0], "to key:", key)


def records_reader():

    # Read in the roads from the CSV file
    with open('data/tiger_roads.csv', newline='') as csvfile:
        reader = csv.reader(csvfile)
        # skip the first line
        next(reader)
        roads = []
        min_max_X = [99999.0, -99999.0]
        min_max_Y = [99999.0, -99999.0]
        
        for i, row in enumerate(reader, start=1):
            # Extract the points from the row
            points = []
            for cord in row:
                cord = cord.split(" ")
                points.append([float(cord[0]), float(cord[1])])
            # points = [[float(p.split()[0]), float(p.split()[1])] for p in row[0].split(',')]

            # Find the min and max x,y points of the MBR
            min_x = 99999.0
            max_x = -9999.0
            min_y = 99999.0
            max_y = -9999.0
            for p in points:
                if p[0] < min_x:
                    min_x = p[0]
                if p[0] > max_x:
                    max_x = p[0]
                if p[1] < min_y:
                    min_y = p[1]
                if p[1] > max_y:
                    max_y = p[1]
            
            #min_x = min(points, key=lambda p: p[0])[0]
            #min_y = min(points, key=lambda p: p[1])[1]
            #max_x = max(points, key=lambda p: p[0])[0]
            #max_y = max(points, key=lambda p: p[1])[1]
            mbr = [[min_x, min_y], [max_x, max_y]]
            
            if min_x < min_max_X[0]:
                min_max_X[0] = min_x
            if max_x > min_max_X[1]:
                min_max_X[1] = max_x
            if min_y < min_max_Y[0]:
                min_max_Y[0] = min_y
            if max_y > min_max_Y[1]:
                min_max_Y[1] = max_y
                
            
            # Save the road data
            road_data = [i, mbr , points]
            roads.append(road_data)
            all_points = ''
            for p in points:
                all_points = all_points + ' '.join(map(str, p)) + ","
            
            # Open the file for writing
            grd = open("data/grid.grd", "a")
            grd.write(str(i) + ", " + str(min_x) + " " + str(min_y) + ", " + str(max_x) + " " + str(max_y) + ", " + str(all_points) + "\n")
            
        grd.close()
                      
    return roads, min_max_X, min_max_Y


records, X, Y = records_reader()

# create grid dictionary
grid = create_grid(X, Y)

add_mbr_to_grid(records, grid)
```
